```python
import numpy as np
x = np.array(([2, 9], [1, 5], [3, 6]), dtype = float)
y = np.array(([92], [86], [89]), dtype = float)
print("Input:\n",x)
print("Output:\n",y)
```

```python
c = np.amax(x, axis = 0)
print(c)
```

```python
x = x/c
y = y/100
print("Normalized Input:\n",x)
print("Normalized Output:\n",y)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def sigmoid_grad(x):
    return x * (1 - x)
epoch = 1000
eta = 0.1
input_neurons = 2
hidden_neurons = 3
output_neurons = 1
```

```python
wh = np.random.uniform(size = (input_neurons, hidden_neurons))
print("Weights input to hidden:\n",wh)
```

```python
bh = np.random.uniform(size = (1, hidden_neurons))
print("Input Bias:\n",bh)
```

```python
wout = np.random.uniform(size = (hidden_neurons, output_neurons))
print("Weights from hidden to Output Layer:\n",wout)
```

```python
bout = np.random.uniform(size = (1, output_neurons))
print("Output Bias:\n",bout)
```

```python
for i in range(epoch):
    print("\n\t( EPOCH {} of {} )".format(i+1, epoch))
    h_ip = np.dot(x, wh) + bh
    print("h_ip:\n", h_ip)
    h_act = sigmoid(h_ip)
    o_ip = np.dot(h_act, wout) + bout
    output = sigmoid(o_ip)
    eo = y - output
    out_grad = sigmoid_grad(output)
    d_output = eo * out_grad
    print("Output:\n", d_output)
    eh = d_output.dot(wout.T)
    hidden_grad = sigmoid_grad(h_act)
    d_hidden = eh * hidden_grad
    wout += h_act.T.dot(d_output) * eta
    wh += x.T.dot(d_hidden) * eta
```
